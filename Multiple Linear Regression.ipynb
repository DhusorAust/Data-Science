```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('G:\personal\EXERCISE\Multiple_regression.csv')
```

```python
X=data.drop(columns=['Profit','State'])
y=data['Profit']

state=data['State']
dummies_state=pd.get_dummies(state)
X=pd.concat([X,dummies_state],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_test
```

```python
y_test_df=pd.DataFrame(y_test)
y_test_df
```

```python
y_prediction=regressor.predict(X_test)

y_prediction_df=pd.DataFrame(y_prediction, columns=['y_predict'])

```

```python
y_prediction_df
```

```python
y_test_df.reset_index(drop=True, inplace=True)
y_prediction_df.reset_index(drop=True, inplace=True)
```

```python
compare=pd.concat([y_test_df,y_prediction_df],axis=1,ignore_index=True,)
compare
```
