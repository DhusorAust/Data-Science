# Aggregation and Grouping

sum() , mean() , median() , min() , and max()

```python
import pandas as pd
import numpy as np
```

```python
rng = np.random.RandomState(42)
df = pd.DataFrame({'A': rng.rand(5),
'B': rng.rand(5)})
df
```

```python
df.mean(axis='columns')
```

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
'data': range(6)}, columns=['key', 'data'])
df1
```

```python
df1.groupby('key').sum()
```

```python
df1.groupby('key').aggregate(['min', np.median, max])
```

```python
rng = np.random.RandomState(0)
df2 = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'data1': range(6),
    'data2': rng.randint(0, 10, 6)},
    columns = ['key', 'data1', 'data2'])
df2
```

```python
df2.groupby('key').aggregate(['min', np.median, max])
```

```python
df2.groupby('key').aggregate({'data1': 'min','data2': 'max'})
```

```python
df3=pd.DataFrame({
    'survived':[0,1,1,1,0],
    'sex':['male','female','female','female','male'],
    'age':[22,33,22,44,24],
    'fare':[7.2500,71.2833,7.9250,53.1000,8.0500],
    'class':['Third','First','Third','First','Third'],
})
df3
```

```python
df3.groupby('sex')[['survived']].mean()
```

```python
df3.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
```

```python
df3.pivot_table('survived', index='sex', columns='class')
```
