```python
import pandas as pd
import numpy as np
```

```python
dataset1=pd.read_excel(r'G:\personal\ML Exercise\Pivot_melting_exteneded.xlsx')
dataset1
```

```python
dataset1_melt=pd.melt(frame=dataset1,id_vars=['Country','Year'])
dataset1_melt
```

```python
dataset1_melt['Gender']=dataset1_melt['variable'].str[0]
dataset1_melt
```

```python
dataset1_melt['Age_Group']=dataset1_melt['variable'].str[1:]
dataset1_melt
```

```python
dataset1_melt.drop(columns='variable',inplace=True,axis=1)
dataset1_melt
```

```python
for i in range(0,len(dataset1_melt)):
    print(dataset1_melt.loc[i, 'Age_Group'])
```

```python
for i in range(0,len(dataset1_melt)):
    temp=dataset1_melt.loc[i, 'Age_Group']
    if len(temp)==3:
        dataset1_melt.loc[i,'Start_Age']=temp[0]
        dataset1_melt.loc[i,'End_Age']=temp[1:]
    else:
        dataset1_melt.loc[i,'Start_Age']=temp[:2]
        dataset1_melt.loc[i,'End_Age']=temp[2:]
        
```
